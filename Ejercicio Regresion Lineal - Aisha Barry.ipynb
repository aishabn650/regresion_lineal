## Root Mean Squared Error (RMSE)

Root Mean Squared Error (RMSE) es una medida de la diferencia entre los valores predecidos y los valores reales. Se calcula tomando la raíz cuadrada de la media de los errores al cuadrado.

El RMSE se utiliza comúnmente como una métrica de evaluación en problemas de regresión, ya que proporciona una idea de qué tan lejos están las predicciones del modelo de los valores reales. Un RMSE más bajo indica un modelo con una mejor precisión en sus predicciones.

Matemáticamente, el RMSE se puede calcular de la siguiente manera:

La fórmula matemática para el RMSE es la siguiente:

$RMSE = sqrt(1/n * sum((y_{real} - y_{pred})^2))$

donde:

- y_real es el valor real o verdadero
- y_pred es el valor predicho por el modelo
- n es el número de observaciones
- sqrt es la función de raíz cuadrada
- sum es la función de suma

En esta fórmula, se suman los cuadrados de la diferencia entre los valores reales y los valores predichos, se divide por el número de observaciones, y luego se toma la raíz cuadrada del resultado. El resultado final es el RMSE, que es una medida de la diferencia media entre los valores reales y los valores predichos.


Aquí está un ejemplo de código en Python para calcular el RMSE:

> En este ejemplo, y_true es una lista con los valores reales, y y_pred es una lista con los valores predichos por el modelo. La función rmse toma ambas listas como argumentos y devuelve el RMSE como resultado.


```python
import numpy as np

def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])
error = rmse(y_true, y_pred)
print(f'RMSE: {error}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_true,y_pred)
```

# Template básico de esquema de regresión

Este es un ejemplo básico de código para introducirte en la regresión en Python utilizando scikit-learn. Puedes utilizar tus propios datos y experimentar con diferentes modelos y métricas para mejorar tus resultados.

1- Importar las librerías necesarias:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

2- Cargar los datos en un DataFrame de Pandas:

```python
data = pd.read_csv("data.csv")
```

3- Dividir los datos en variables predictoras y variable objetivo:

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

4- Crear el modelo de regresión lineal:

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

5- Hacer predicciones con el modelo:

```python
y_pred = regressor.predict(X)
```

6- Calcular el error cuadrático medio (MSE):

```python
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```

7- Visualizar los resultados de la regresión:

```python
plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='blue')
plt.show()
```
